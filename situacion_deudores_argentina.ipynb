# Transformaciones

```python
df1=df.copy()
```

```python
# Eliminar columna 2
df1 = df1.drop(columns=2)
```

```python
df1.head(10)
```

```python
# Asignar situación a los códigos de línea

df1["situacion"] = (
    df1[3]
    .astype(str)
    .str[-2:]  # últimos dos dígitos
    .map({
        "10": "Situacion 1",
        "20": "Situacion 2",
        "30": "Situacion 3",
        "40": "Situacion 4",
        "50": "Situacion 5"
    })
)
```

```python
# Verificar cambios realizados

df1.head(3)
```

```python
# Cambiar nombre a las columnas

df1 = df1.rename(columns={
    0: "codigo_entidad",
    1: "nombre_entidad",
    3: "codigo_linea",
    4: "denominacion",
    5: "dic_2023",
    6: "dic_2024",
    7: "set_2025",
    8: "oct_2025",
    9: "nov_2025",
    10: "formato"
})
```

```python
# Verificar cambios realizados

df1.columns
```

```python
cols_valor = [
    "dic_2023",
    "dic_2024",
    "set_2025",
    "oct_2025",
    "nov_2025"
]
```

```python
df1 = df1.melt(
    id_vars=[
        "codigo_entidad",
        "nombre_entidad",
        "codigo_linea",
        "denominacion",
        "formato",
        "situacion"
    ],
    value_vars=cols_valor,
    var_name="periodo",
    value_name="valor"
)
```

```python
df1["periodo"] = (
    df1["periodo"]
    .str.replace("valor_", "", regex=False)
)
```

```python
# Verificar cambios realizados

df1.head(3)
```

```python
# Convertir a tipo fecha

map_fecha = {
    "dic_2023": "2023-12-31",
    "dic_2024": "2024-12-31",
    "set_2025": "2025-09-30",
    "oct_2025": "2025-10-31",
    "nov_2025": "2025-11-30"
}

df1["fecha"] = pd.to_datetime(df1["periodo"].map(map_fecha))
```

```python
df1.head(3)
```

```python
# Ordenar columnas

df1 = df1[
    [
        "codigo_entidad",
        "nombre_entidad",
        "codigo_linea",
        "denominacion",
        "situacion",
        "fecha",
        "periodo",
        "valor",
        "formato"
    ]
]
```

```python
# Verificar cambios realizados

df1
```

```python
# Reemplazar los valores nan en la columna situación por total
df1['situacion'] = df1['situacion'].fillna('total')
```

```python
# Cambiar valores de la columna formtao
df1['formato'] = df1['formato'].map({
    'N': 'monto',
    'P': 'porcentaje'
})
```

```python
df1
```

```python
# Eliminar registros que no corresponden a entidades financieras
df1 = df1[~df1['codigo_entidad'].str.startswith('AA', na=False)]
```

```python
# Verificar cantidad de filas despues de la eliminación
print("Filas después:", len(df1))
```

```python
# Eliminar columnas codigo_linea y periodo
df1 = df1.drop(columns=['codigo_linea', 'periodo'], errors='ignore')
```

```python
# Verificar eliminación de columnas
df1.columns
```

```python
df1.head(30)
```

```python
# Eliminar registros donde denominación sea igual a "Total cartera sobre financiaciones (%)""

df1 = df1[df1["denominacion"] != "Total cartera sobre financiaciones (%)"].copy()
```

```python
df1.head(30)
```

```python
# Eliminar registros donde aparece: "TOTAL GARANTIZADO|Previsiones por riesgo de incobrabilidad"

mask_excluir = df1["denominacion"].str.contains(
    r"TOTAL GARANTIZADO|Previsiones por riesgo de incobrabilidad",
    case=False,
    na=False
)

df1 = df1[~mask_excluir].copy()
```

```python
df1.head(30)
```

```python
# Eliminar registros donde aparece: "^TOTAL DE FINANCIACIONES|^TF\.Sit"

mask_tf = df1["denominacion"].str.contains(
    r"^TOTAL DE FINANCIACIONES|^TF\.Sit",
    case=False,
    na=False
)

df1 = df1[~mask_tf].copy()
```

```python
df1.head(30)
```

```python
df1["cartera"] = np.select(
    [
        # orden IMPORTANTE (de más específico a más general)
        df1["denominacion"].str.contains(r"C\.CAC|ASIMILABLE A CONSUMO", case=False, na=False),
        df1["denominacion"].str.contains(r"C\.CON|CONSUMO O VIVIENDA", case=False, na=False),
        df1["denominacion"].str.contains(r"C\.COM|CARTERA COMERCIAL", case=False, na=False),
    ],
    [
        "comercial_asimilable_consumo",
        "consumo_vivienda",
        "comercial",
    ],
    default="sin_clasificar"
)
```

```python
df1.head(30)
```

```python
df1
```

```python
# Separar montos y porcentaje

df_montos = df1[df1["formato"] == "monto"].copy()
df_porcentajes = df1[df1["formato"] == "porcentaje"].copy()
```

```python
# Verificar tipo númerico
df_porcentajes["valor"] = pd.to_numeric(df_porcentajes["valor"], errors="coerce")
df_montos["valor"] = pd.to_numeric(df_montos["valor"], errors="coerce")
```

```python
# Normalizar montos
df_montos["valor"] = df_montos["valor"] / 100
```

```python
# Obtener el total de cada cartera para multiplicar los porcentajes 
df_totales = df_montos[df_montos["situacion"] == "total"].copy()
```

```python
# nos quedamos solo con lo necesario
df_totales = df_totales[["codigo_entidad", "fecha", "cartera", "valor"]].rename(columns={"valor": "monto_total"})
```

```python
# Combinar con df_porcentajes para calcular valores nominales 
df_porcentajes = df_porcentajes.merge(df_totales, on=["codigo_entidad", "fecha", "cartera"], how="left")
```

```python
#Calcular valor nominal 
df_porcentajes["valor"] = (df_porcentajes["valor"]/100 * df_porcentajes["monto_total"])
df_porcentajes["formato"] = "monto"
```

```python
# Dataset final
df_final = df_porcentajes.drop(columns=["monto_total"]).copy()
```

```python
# Verificar cambios realizados
df_final.head(30)
```

```python
# Eliminar columnas innecesarias
df_final = df_final.drop(columns=["denominacion", "formato"])
```

```python
# Reordenar columnas
df_final = df_final[["codigo_entidad", "nombre_entidad", "cartera", "situacion", "fecha", "valor"]]
```

```python
# Verificar cambios realizados
df_final.head(10)
```
